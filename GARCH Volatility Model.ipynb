```python
import numpy as np
from matplotlib.pyplot import *

v = 850 # number of observations
v1 = 250 # we need to drop the first several observations
v2 = v + v1 # total number of observations
alpha0 = 0.2 # GARCH(1,1) coefficient alpha0
alpha1 = 0.4 # GARCH(1,1) coefficient alpha1
beta1 = 0.3 # GARCH(1,1) coefficient beta1
errors = np.random.normal(0, 1, v2)
t1 = np.zeros(v2)
t1[0] = np.random.normal(0, np.sqrt(alpha0 / (1 - alpha1 - beta1)), 1)
for i in range(1, v2):
    t1[i] = errors[i] * np.sqrt(alpha0 + alpha1 * t1[i-1]**2 + beta1 * t1[i-1]**2)

y = t1[v1:]
x = range(v)
plot(x, y)
title('GARCH(1,1) process')
show()
```

```python
import yfinance as yf
import scipy as sp
import matplotlib.pyplot as plt
import scipy import random
import numpy as np

# Download daily data for S&P 500 from Yahoo Finance
symbol = "^GSPC"
start_date = "2015-01-01"
end_date = "2022-04-28"
data = yf.download(symbol, start=start_date, end=end_date)

# Compute daily log returns
returns = sp.log(data["Adj Close"] / data["Adj Close"].shift(1))

# Compute GARCH(1,1) volatility
v = len(returns)
v1 = int(v * 0.1)
v2 = v + v1
alpha0 = 0.2
alpha1 = 0.4
beta1 = 0.3
errors = sp.random.normal(0, 1, v2)
t1 = sp.zeros(v2)
t1[0] = sp.random.normal(0, sp.sqrt(alpha0 / (1 - alpha1 - beta1)), 1)
for i in range(1, v2):
    t1[i] = errors[i] * sp.sqrt(alpha0 + alpha1 * t1[i-1]**2 + beta1 * t1[i-1]**2)
volatility = t1[v1:]

# Plot S&P 500 log returns and GARCH(1,1) volatility
fig, ax1 = plt.subplots()
color1 = "tab:red"
ax1.set_xlabel("Date")
ax1.set_ylabel("Log Returns", color=color1)
ax1.plot(returns.index, returns, color=color1)
ax1.tick_params(axis="y", labelcolor=color1)
ax2 = ax1.twinx()
color2 = "tab:blue"
ax2.set_ylabel("Volatility", color=color2)
ax2.plot(returns.index, volatility, color=color2)
ax2.tick_params(axis="y", labelcolor=color2)
plt.title("S&P 500 Log Returns and GARCH(1,1) Volatility")
plt.show()
```

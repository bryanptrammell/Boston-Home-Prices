```python
import statsmodels.api as sm
import pandas as pd

def regress(df, df3):
    # add a constant column to df2
    df3 = sm.add_constant(df3)
    # combine df1 and df3
    combined_df = pd.concat([df, df3], axis=1)
    # fit an OLS model
    model = sm.OLS(combined_df.iloc[:,0], combined_df.iloc[:,1:])
    results = model.fit()
    # print the summary statistics
    print(results.summary())
    return results
```
